**Unit Price:** The cost per individual item.

**Total Sale:** The total monetary value of the order.

**Sales Channel:** Where the purchase was made (e.g., Online, Wholesale, or In-Store).

# Stage 2 – Data Cleaning and Pre-processing

```python
import numpy as np # Ensure numpy is imported for np.float64
# Impute missing values column with mean statistics because here filled with mode with own category and mean for numerical statistics along with change data type.
df['Customer Name'] = df['Customer Name'].fillna("unknown")
# First, convert to standard float type, then fill NaNs, and then ensure it's float64
df['Units Sold'] = df['Units Sold'].astype(np.float64).fillna(df['Units Sold'].mean()).astype(np.float64)
df['Unit Price'] = df['Unit Price'].fillna(df['Unit Price'].mean())
df['Total Sale'] = df['Total Sale'].fillna(df['Total Sale'].mean())
df['Sales Channel'] = df['Sales Channel'].fillna("unknown")
df['Order ID'] = df['Order ID'].fillna("unknown")
df.isnull().sum()
```

```python
# Duplicated used to check whether rows are repeated (duplicated rows) in dataframe
df.duplicated().sum()
```

```python
#  Impute missing values with statistics of mode or own categorical columns
df['Sales Channel'] = df['Sales Channel'].fillna("unknown")
df['Customer Name'] = df['Customer Name'].fillna("unknown")
df
```

```python
# Finding profit total sale and unit price # Feature transformations (date parts, derived fields if required for analysis)
df['Profit'] = df['Total Sale'] / df['Unit Price']
df
```

```python
columns = ["Units Sold","Unit Price","Profit"]
df[columns].skew()
```

```python
# Feature transformations (date parts, derived fields if required for analysis)
df['discount Margin'] = df['Total Sale'] / df['Profit'] * 100
df
```

```python
df['Sale Date'] = pd.to_datetime(df['Sale Date'],dayfirst=0)
df
```

```python
# Handle missing values (impute or drop)
df.drop_duplicates(subset=['Order ID'])
df
```

```python
import matplotlib.pyplot as plt

# Boxplot
plt.boxplot([df['Unit Price'],df['discount Margin']],patch_artist=True,boxprops=dict(color='red'),medianprops=dict(color='blue'),whiskerprops=dict(color='purple'),capprops=dict(color='green'))
plt.title('Distribution of discount margin by unit price')
plt.xlabel('Unit Price')
plt.ylabel('Discount Margin')
plt.show()
# Detect Outliers Using IQR Method
Q1_up = df['Unit Price'].quantile(0.25)
Q3_up = df['Unit Price'].quantile(0.75)

IQR_up = Q3_up - Q1_up

lower_up = Q1_up - 1.5 * IQR_up
upper_up = Q3_up + 1.5 * IQR_up
Q1_dm = df['discount Margin'].quantile(0.25)
Q3_dm = df['discount Margin'].quantile(0.75)

IQR_dm = Q3_dm - Q1_dm

lower_dm = Q1_dm - 1.5 * IQR_dm
upper_dm = Q3_dm + 1.5 * IQR_dm
# Treat Outliers (Capping Method)
df.loc[:, 'Unit Price'] = df['Unit Price'].clip(lower_up, upper_up)

df.loc[:, 'discount Margin'] = df['discount Margin'].clip(lower_dm, upper_dm)
# After Treatment

plt.figure(figsize=(8,6))

plt.boxplot(
    [df['Unit Price'], df['discount Margin']],
    patch_artist=True,
    boxprops=dict(color='red'),
    medianprops=dict(color='blue'),
    whiskerprops=dict(color='purple'),
    capprops=dict(color='green')
)

plt.title('Distribution of Unit Price and Discount Margin')
plt.xticks([1,2], ['Unit Price', 'Discount Margin'])

plt.ylabel('Values')

plt.show()
```

```python
# Univariate Analysis → distribution of single variables (countplot, histogram, boxplot)
import seaborn as sns
import pandas as pd
plt.figure(figsize=(20, 10))
sns.countplot(x='Product',data=df)
plt.title('count of sales by product')
plt.show()
```

**Count Plot Interpretation:**
This countplot reveals the popularity of product categories. If electronics are the most frequently sold, it signals high consumer demand, which should be aligned with inventory management to avoid stockouts.

```python
# HISTOGRAM
sns.histplot(['Total Sale'],bins=20,kde=True)
plt.title('Total Sale in Nigerian State')
plt.show()
```

**Histogram Interpretation:**
This histogram highlights the spread of Sale.
A right-skewed distribution indicates that a small number of transactions generate a disproportionately high amount of Sale, suggesting a need to focus on high-value customers.

**Box Plot Interpretation:** The boxplot identifies transactions that are significantly higher or lower than the median. Outliers on the higher side represent potential bulk purchases, crucial for identifying B2B clients in the wholesale channel.

```python
# Bivariate Analysis → relation between two variables (scatterplot, barplot, correlation heatmap)
# Scatter Plot
plt.scatter(df['Unit Price'],df['Total Sale'],color='red',alpha=0.7,edgecolor='green')
plt.xlabel('Unit Price')
plt.ylabel('total Sale')
plt.title('unit Price vs Total Sale')
plt.grid(True)
plt.show()
```

**Scatter Plot Interpretation:** This scatterplot shows a positive correlation, indicating that higher-priced items generally lead to higher revenue. Regional color-coding reveals if premium products (high price) are dominating in specific areas like Lagos or Abuja.

```python
# Bar Plot
total = df.groupby('Sales Channel')['Total Sale'].sum()
plt.bar(total.index,total.values,color='lightblue')
plt.xlabel('Sales Channel')
plt.ylabel('Total Sale')
plt.xticks(rotation=45)
plt.title('Total Sale by Sales channel')
plt.show()
```

**Bar Plot Interpretation:** The barplot shows which sales channel—Wholesale, Retail, or Online—contributes most. If wholesale outperforms online, it confirms that bulk B2B strategies are more effective for revenue growth in the current landscape.

```python
# correlation Heatmap
correlation=df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.show()
```

**HeatMap Interpretation:** This heatmap highlights strong correlations (e.g., strong positive link between UnitsSold and TotalRevenue). A low correlation between ProductDiscount and Profit might indicate that current discounts are not effectively driving profitability.

```python
# Multivariate Analysis → relation among 3+ variables (pairplot, grouped analysis, pivot tables, advanced plots)
#pair plot
sns.pairplot(df,vars=['Units Sold','Unit Price','Total Sale'],hue='State')
plt.show()
```

**Pair Plot Interpretation:**
The pairplot serves as a summary
It shows the distribution of every metric while highlighting regional clusters.
If one region’s data points (e.g., North-Central) are consistently clustered at the lower end of the UnitsSold vs TotalSale scatter, it signals a market that is either underserved or lacks purchasing power, requiring a specialized pricing intervention.

```python
#Pivot Table
pivot = pd.pivot_table(
    df,
    values='Total Sale',
    index='State',
    columns='Product',
    aggfunc='sum'
)

print(pivot)
plt.figure(figsize=(20,10))

sns.heatmap(pivot.fillna(0).astype(float), annot=True, cmap='YlGnBu')

plt.title("Sales by State and Product")
plt.show()
```

**Pivot table Interpretation:** This visualization is critical for resource allocation. It identifies "Dead Zones" (low profit/certain channel) and "Gold Mines" (high profit/certain channel). If Home Appliances show higher margins through Wholesale than Retail, the sales team should pivot their negotiation efforts toward bulk B2B contracts for that category.

Thick Black Bar: Represents the Interquartile Range (IQR), containing the middle 50% of the data.

Thin Black Line: Represents the whiskers, showing the rest of the distribution range, excluding extreme outliers.

```python
# bubble Plot
import plotly.express as px
data = px.scatter(df,
                  x='Units Sold',
                  y='Unit Price',
                  size='Total Sale',
                  color='Profit',
                  title='Profit Margin',
                  width=1500,   # Controls width in pixels
                  height=500)   # Controls height in pixels
data.show()
```
